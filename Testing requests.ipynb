```python
#https://americas.api.riotgames.com/lol/match/v5/matches/BR1_2867294688?api_key=RGAPI-26e8d50f-3044-4d2d-bd6d-66a5d9743e41

def getMatchData(region, matchId, apiKey):
    apiUrl = (
        "https://" +
        region +
        ".api.riotgames.com/lol/match/v5/matches/" +
        matchId +
        "?api_key=" + 
        apiKey
    )
    
    #print(apiUrl)
    resp = requests.get(apiUrl)
    return resp.json()
```

```python
region = 'americas'
matchId = 'BR1_2865467209'
apiKey = 'RGAPI-26e8d50f-3044-4d2d-bd6d-66a5d9743e41'
```

```python
def didWin(puuid, matchData):
    participants = matchData['metadata']['participants']
    partIndex = participants.index(puuid)
    #return match['info']['participants'][partIndex]
    part = matchData['info']['participants'][partIndex]
    print(part['championName'])
    return part['win']
```
